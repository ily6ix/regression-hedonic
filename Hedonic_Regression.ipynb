# Hedonic Regression: Housing Price Prediction

This notebook demonstrates a **hedonic regression model** to predict housing prices based on key features:
- Size (in square meters)
- Number of bedrooms
- Distance from city center (km)

We will use **Python**, **pandas**, and **scikit-learn** for the regression, and **matplotlib / seaborn** for visualization.


## What is Hedonic Regression?

A **hedonic regression** estimates the price of a product (like a house) by evaluating the contribution of its individual features.

Mathematically, it is represented as:

House Price = β₀ + β₁*(Size) + β₂*(Bedrooms) + β₃*(Distance) + ε

Where:
- **β₀ (Intercept)** = base price of the house
- **β₁, β₂, β₃** = effect of each feature on price
- **ε** = error term (difference between predicted and actual prices)

This model helps answer questions like:
- How much does an extra bedroom add to the price?
- How much does distance from the city reduce the price?

```python

# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots appear in the notebook
%matplotlib inline
```

```python

# Sample housing dataset
data = {
    "size": [50, 80, 100, 120, 150],
    "bedrooms": [1, 2, 3, 3, 4],
    "distance": [10, 7, 5, 4, 2],
    "price": [450000, 750000, 950000, 1150000, 1500000]
}

df = pd.DataFrame(data)
df
```

## Select Features and Target

- **Features (X):** input variables that influence price (size, bedrooms, distance)  
- **Target (y):** the output variable we want to predict (price)


```python

# Define features and target
X = df[["size", "bedrooms", "distance"]]
y = df["price"]
```

## Train Linear Regression Model

- Create a Linear Regression model
- Fit the model on our dataset
- The model learns coefficients (β values) for each feature


```python
# Initialize and train the model
model = LinearRegression()
model.fit(X, y)

# Display learned coefficients
print("Intercept (base price):", model.intercept_)
print("Coefficients (effect of features):", model.coef_)

```

## Predict Price of a New House

We can use the trained model to predict the price of a new house by passing a **DataFrame** with the same feature names:

Example:
- Size: 100 m²
- Bedrooms: 3
- Distance from city: 5 km



```python
# Create a new house DataFrame
new_house = pd.DataFrame([[100, 3, 5]], columns=["size", "bedrooms", "distance"])

# Predict price
predicted_price = model.predict(new_house)
print("Predicted Price:", predicted_price[0])
```

## Visualize Relationship Between Features and Price

- Scatter plots help us see how each feature affects the price
- Positive slope → feature increases price
- Negative slope → feature decreases price


```python
sns.pairplot(df, x_vars=["size","bedrooms","distance"], y_vars="price", height=4, kind="scatter")
plt.show()
```

# Summary

- Hedonic regression predicts **house prices** based on features.
- Linear regression estimates the **value contribution** of each feature.
- Model predictions can be made for new data points.
- Visualizations help understand relationships and validate the model.
